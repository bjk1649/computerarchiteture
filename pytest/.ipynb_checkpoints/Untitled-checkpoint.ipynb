```python
def train_test_split(X,Y,random=True,randomSeed=None,testSize=0.2):
    #인덱스 생성
    index=  np.arange(X.shape[0])
    size = round(testSize * X.shape[0])
    if random ==True:
        if randomSeed !=None :
            randomGene = np.random.RandomState(seed=randomSeed)
            randomGene.shuffle(index)
        
        testIndex=index[:size]
        trainIndex=index[size:]
        XTest= X.iloc[testIndex,:]
        YTest=Y.iloc[testIndex]
        XTrain = X.iloc[trainIndex,:]
        YTrain = Y.iloc[trainIndex]

            
    return XTest, YTest, XTrain,YTrain
    
```
