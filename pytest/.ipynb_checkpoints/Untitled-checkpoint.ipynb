```python
def train_test_split(X,Y,random=True,randomSeed=None,testSize=0.2):
    #인덱스 생성
    index=  np.arange(X.shape[0])
    size = round(testSize * X.shape[0])
    if random ==True:
        if randomSeed !=None :
            randomGene = np.random.RandomState(seed=randomSeed)
            randomGene.shuffle(index)
        
        testIndex=index[:size]
        trainIndex=index[size:]
        XTest= X.iloc[testIndex,:]
        YTest=Y.iloc[testIndex]
        XTrain = X.iloc[trainIndex,:]
        YTrain = Y.iloc[trainIndex]

            
    return XTest, YTest, XTrain,YTrain
    
```

```python
def EuclideanDistance (X1, X2):
    
    distance = (((X1[0]-X2[0])**2)+((X1[1]-X2[1])**2)+((X1[2]-X2[2])**2)+((X1[3]-X2[3])**2))**1/2
    ## distance = np.sqrt(np.sum((X1-X2)**2))
    
    return distance
```

```python
EuclideanDistance(obs1,obs2)
```

```python
def KNN (X, Xtest, k=3):
    
    temp = np.array(X)
    distanceList = []
    indexList = []
    
    for i in range(len(X)):
        distance = EuclideanDistance(temp[i,:],Xtest)
        distanceList.append((distance,i))
        
    distanceList.sort()
    for j in range(k):
        indexList.append(distanceList[j][1])
    
    return indexList
```

```python
def prediction (Train_X, Train_Y, test):
    index = KNN(Train_X , test , 3)
    
    result =[]
    testli=Ytrain.values.tolist()
    for j in index:
        result = testli[j]
    
    result2 = np.array(result)
    #unique, counts = np.unique(result2, return_counts=True)
    #rlist = dict(zip(unique,counts))
    
    
    return result2
```
