#### calc 함수 정의

```python
def calc(n):
    sum = 0
    for i in range(n) :
        sum += int(input())
    
    return sum
```

#### 함수 사용

```python
print('정수의 갯수를 입력하세요')
n = int(input())
while n<=0 :
    print('정수의 갯수를 입력하세요')
    n = int(input())

sum = calc(n)

print('결과 : {:.2f}'.format(sum))
```

```python
def gauss(A):
    (n,m) = A.shape
    for i in range(0,n):
        maxEl = abs(A[i,i])
        maxRow = i
        
        for j in range(i+1,n):
            if abs(A[j,i])>maxEl:
                maxRow = j
                maxEl = abs(A[j,i])
        
        for j in range(i,m):
            tmp = A[i,j]
            A[i,j] = A[maxRow,j]
            A[i,j] = tmp
        
        piv = A[i,i]
        for j in range(i,m):
            A[i,j] = A[i,j]/A[i,i]
        
        for j in range(0,n):
            if j !=i:
                c = A[j,i]/A[i,i]
                
                
                for k in range(i,m):
                    if i == k:
                        A[j,k] = 0
                    else : 
                        A[j,k] = A[j,k] - c*A[i,k]
                
        pprint(str(i+1)+"번째 반복 ", A)
        
    
    x = np.zeros(n)
    for i in range(n):
        x[i] = A[i,n]
    return x
                
```

```python
A = np.array([[2.,2.,4.,18],[1.,3.,2.,13.],[3.,1.,3.,14.]])
```

```python
def pprint(msg,A):
    print ("---",msg,"---")
    (n,m) = A.shape
    for i in range(n):
        line =""
        for j in range(m):
            line += "{:.2f}".format(A[i,j])+"\t"
            if j == m-2:
                line+="| "
        print(line)
    print("")
```
