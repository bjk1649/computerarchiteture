```python
class MLR_class:
    

    def intercept_add(self,X):
        ones = np.ones(len(X))
        ones.reshape(-1,1)
        trueX = np.concatenate([ones,X],axis=1)
        return trueX
    
    def fit(self,X,Y,intercept=False):
        if intercept==True:
            X= self.intercept_add(X)
        coef = np.linalg.inv(X.transpose()@X)@X.transpose()@Y
        return coef
    
    def predict(self,X,coef,intercept=False):
        if intercept ==True:
            X= self.intercept_add(X)
        result = X@coef
        return result
    
```
