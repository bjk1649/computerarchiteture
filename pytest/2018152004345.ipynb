```python
Y = new_data['Life expectancy'] # 6번 수명기댓값을 Y로 지정, 나머지를 X로 지정
X = new_data.drop(['Life expectancy'],axis=1)
```

```python
X=np.array(X)
Y=np.array(Y)

X_train = X[0:137,:] #6번 X,Y를 137개 기준으로 트레인, 테스트로 나눔 
Y_train = Y[0:137]
X_test = X[137:,:]
Y_test = Y[137:]
```

```python
class MLR: #7번 MLR클래스 정의
    def __init__(self,intercept =True):
        self.intercept = intercept
        
        
    def intercept_add(self,X):
        new_ones = np.ones(len(X))
        plus_data = new_ones.reshape(-1,1)
        new_X = np.concatenate([X,plus_data],axis=1)
        
        return new_X
    
    def fit(self,X,Y):
        if self.intercept==True:
            X = self.intercept_add(X)
            
        self.coef = np.linalg.inv(X.transpose()@X)@X.transpose()@Y
        return self.coef
    
    def predict(self,X):
        predicted = list()
        
        for sample_x in X:
            if self.intercept==True:
                sample_x=self.intercept_add(sample_x)
        
            Y_pred = sample_x@self.coef
            predicted.append(Y_pred)
            
        predicted = np.array(predicted)
        return predicted
```

```python
class KNN: #8번 knn 클래스 정의
    def __init__(self,n_neighbor=5):
        self.n_neighbor =5
        self.neighbors = None
    
    def euclidean_dist(self,obs1,obs2):
        return np.sqrt(np.sum((obs1-obs2)**2))
    
    def search_neighbors(self,X_train,test_sample_x):
        dists_info = list()
        for index, train_sample_x in enumerate(X_train):
            dist = self.euclidean_dist(train_sample_x,test_sample_x)
            dists_info.append((dist,index,train_sample_x))
            
        dists_info.sort(key=lambda x : x[0])
        ne_in = [ne[1] for ne in dists_info]
        return ne_in[:self.n_neighbor]
    
    def predict(self, X_train, Y_train, X_test):
        predicted = list()
        for test_sample_x in X_test:
            index = self.search_neighbors(X_train,test_sample_x)
            predicted.append(np.mean(Y_train[index]))
            
        return predicted
```

```python
def MSE(true_y,pred_y): # 9번 MSE, MAPE 함수 정의
    result = (np.sum((true_y-pred_y)**2))/len(true_y)
    return result

def MAPE(true_y, pred_y):
    result = np.sum(np.abs(true_y-pred_y))/len(true_y)
    return result
    
```

```python
mlr = MLR(False) #MRL 활용하여 지표측정
mlr.fit(X_train,Y_train)
pred_y2=mlr.predict(X_test)
result3 = MSE(Y_test,pred_y2)
result4 = MAPE(Y_test,pred_y2)
```

```python
knn = KNN(5) # knn 활용하여 지표측정
pred_y=knn.predict(X_train,Y_train,X_test)
pred_y= np.array(pred_y)
result1 = MSE(Y_test,pred_y)
result2 = MAPE(Y_test,pred_y)
```

```python
print('MSE 지표 : Knn : {:.1f} vs MLR : {:.1F}\nMAPE 지표 : {:.1f} vs MLR : {:.1F}'.format(result1,result3,result2,result4)) # 10번 출력
```
