```python
def pprint(msg,A):
    print("---",msg,"---")
    (n,m) = A.shape
    for i in range(n):
        line = ""
        for j in range(m):
            line += "{:.2f}".format(A[i,j]) + "\t"
        print(line)
    print("")
    
```

```python
def LU(A):
    (n,m) = A.shape
    L= np.zeros((n,n))
    U= np.zeros((n,n))
    for i in range(n):
        for j in range(i,n):
            U[i,j] = A[i,j]
            for k in range(i):
                U[i,j] = A[i,j] - L[i,k]*U[k,j]
                print("U",i+1,j+1,"- L",i+1,k+1,"*U",k+1,j+1)
            pprint("U",U)
        
        L[i,i] = 1
        pprint("L",L)
        if i<n-1:
            p = i+1
            for j in range(p):
                L[p,j] = A[p,j]
                for k in range(j):
                    L[p,j] = L[p,j] - L[p,k]*U[k,j]
                L[p,j] = L[p,j]/ U[j,j]
                pprint("L",L)
    
    return L,U
        
        
    
```

```python
A = np.array([[5.,3.,2.,1.],[6.,2.,4.,5.],[7.,4.,1.,3.],[4.,3.,5.,2.]])
```

```python
L,U = LU(A)
```
