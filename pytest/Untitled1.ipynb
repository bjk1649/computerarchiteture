```python
def train_test_split(X,Y,test_size=0.2,validation_size=0.2, random=False, random_seed = None):
    test_size_len = round(test_size *len(X))
    validation_size_len = round(validation_size*len(X))
    indices = np.array(range(len(X)))
    X= np.array(X)
    Y= np.array(Y)
    if random==True:
        
        if random_seed !=None:
            random_generator = np.random.RandomState(random_seed)
            random_generator.shuffle(indices)
            
            test_indices = indices[0:test_size_len]
            validation_indices = indices[test_size_len:test_size_len+validation_size_len]
            train_indices = indices[test_size_len+validation_size_len:]
            
        else :
            np.random.shuffle(indices)
            test_indices = indices[0:test_size_len]
            validation_indices = indices[test_size_len:test_size_len+validation_size_len]
            train_indices = indices[test_size_len+validation_size_len:]
        
        X_train = X[train_indices,:]
        Y_train = Y[train_indices]
        X_test = X[test_indices, :]
        Y_test = Y[test_indices]
        X_validation= X[validation_indices,:]
        Y_validation= Y[validation_indices]
    
    else :
        X_train = X[0:len(X)-test_size_len-validation_size_len,:]
        Y_train = Y[0:len(X)-test_size_len-validation_size_len]
        X_test = X[len(X)-test_size_len-validation_size_len:len(X)-validation_size_len,:]
        Y_test = Y[len(X)-test_size_len-validation_size_len:len(X)-validation_size_len]
        X_validation= X[len(X)-validation_size_len,:]
        Y_validation= Y[len(X)-validation_size_len:]
        
    return X_train, Y_train, X_test, Y_test, X_validation, Y_validation
        
```

```python
def euclidean_dist(obs1,obs2):
    return np.sqrt(np.sum((obs1-obs2)**2))
    
```

```python
def manhattan_dist(obs1,obs2):
    return np.sum(np.abs(obs1-obs2))
```

```python
def search_neighbors(X_train, test_sample,k=5):
    dists_info = list()
    
    for index, train_sample in enumerate(X_train):
        dist = euclidean_dist(test_sample,train_sample)
        dists_info.append((dist,index,train_sample))
        
    dists_info.sort()
    
    return dists_info[:k]
```

```python
def search_neighbors2(X_train,test_sample,k=5):

    dists_info = list()
    
    for index,train_sample in enumerate(X_train):
        dist = manhattan_dist(train_sample,test_sample)
        dists_info.append((dist,index,train_sample))
        
    dists_info.sort(key=lambda x : x[0])
    
    return dists_info[:k]
```

```python
class Knn :
    def __init__ (self, k=5):
        self.k = k
        self.neighbors =None
    
    def euclidean_dist (self, obs1, obs2):
        return np.sqrt(np.sum((obs1-obs2)**2))
    
    def search_neighbors (self, X_train, test_sample):
        dists_info = list()
        for index, train_sample in enumerate(X_train):
            dist = self.euclidean_dist(train_sample,test_sample)
            dists_info.append((dist,index,train_sample))
            
        dists_info.sort(key=lambda x:x[0])
        self.neighbors = dists_info[0:self.k]
        return self.neighbors
    
    def predict(self, X_train, test_sample, Y_train):
        pred_set = list()
        self.neighbors =  self.search_neighbors(X_train,test_sample)
        neigh_index = [info[1] for info in self.neighbors]
        
        neigh_Y = Y_train[neigh_index]
        
        classes, counts = np.unique(neigh_Y,return_counts=True)
        pred_Y = classes[np.argmax(counts)]
        return pred_Y
```
