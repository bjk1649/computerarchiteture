```python
def gauss(A):
    
    (n,m)= A.shape
    for i in range(n):
        maxRow = i
        maxEl = abs(A[i][i])
        
        for j in range(i+1,n):
            if abs(A[j][i])>maxEl:
                maxEl = abs(A[j][i])
                maxRow = j
        
        for j in range(m):
            tmp = A[i][j] 
            A[i][j] = A[maxRow][j]
            A[maxRow][j] = tmp
        
        piv = A[i][i]
        for j in range(m):
            A[i][j] = A[i][j]/piv
            
        for j in range(n):
            if j!=i:
                c = A[j][i]
                for k in range(i,m):
                    A[j][k] = A[j][k] - c*A[i][k]
        pprint(str(i+1) + "번쨰 반복",A )
    x=np.zeros(n)
    for i in range(n):
        x[i] = A[i,n]
    
    return x
                    
```

```python
A = np.array([[2.,2.,4.,18.],[1.,3.,2.,13.],[3.,1.,3.,14.]])
```

```python
def pprint(msg,A):
    print("---",msg,"---")
    (n,m) = A.shape
    
    for i in range(n):
        line = ""
        
        for j in range(m):
            line += "{0:.2f}".format(A[i][j])+"\t"
            if j == n-1 :
                line+="| "
        print(line)
    print("")
```
